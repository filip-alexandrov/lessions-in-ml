```python
from transformers import pipeline, set_seed

import textwrap 
import numpy as np
import matplotlib.pyplot as plt 

from pprint import pprint
```

```python
!head robert_frost.txt
```

```python
lines = [line.rstrip() for line in open('robert_frost.txt')]
lines = [line for line in lines if len(line) > 0]
```

```python
gen = pipeline("text-generation")
```

```python
set_seed(1234)
```

```python
lines[0]
```

```python
# generates additional text
gen(lines[0])
```

```python
pprint(_)
```

```python
pprint(gen(lines[0], max_length = 20))
```

```python
pprint(gen(lines[0], num_return_sequences = 3, max_length = 20))
```

```python
def wrap(x): 
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True) 
```

```python
out = gen(lines[0], max_length = 60)
print(wrap(out[0]['generated_text']))
```

```python
prev = "Two roads diverged in a yellow wood, then the bridge collapsed, as there was little resistance."

out = gen(prev + '\n' + lines[2], max_length = 60)
print(wrap(out[0]['generated_text']))
```

```python
prev = """
Two roads diverged in a yellow wood, then the bridge collapsed, as
there was little resistance.
And be one traveler, long I stood upon
this bleak and dangerous road in our midst, at the mouth of the
mountain and of the mountains and across all my territory and life
there.  
"""

out = gen(prev + "\n" + lines[4], max_length = 90) 
print(wrap(out[0]['generated_text']))
```

```python
prompt = "Neural networks with attention have been used with great success in natural language processing. "
out = gen(prompt, max_length = 300) 
print(wrap(out[0]['generated_text']))
```
