```python
import json
import yt_dlp

URL = 'https://www.youtube.com/watch?v=BaW_jenozKc'

# ‚ÑπÔ∏è See help(yt_dlp.YoutubeDL) for a list of available options and public functions
ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([URL])


```

```python
import yt_dlp as youtube_dl

# Set the YouTube video URL and video ID
video_id = 'Hwybp38GnZw'
video_url = f'https://www.youtube.com/watch?v={video_id}'

# Set the output filename
output_template = f'{video_id}.%(ext)s'

# Create a YouTubeDL object with audio-only format options
ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': output_template,
}

# Create a downloader object with the specified options
ydl = youtube_dl.YoutubeDL(ydl_opts)

# Download the audio
ydl.download([video_url])

```
