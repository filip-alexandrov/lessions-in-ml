```python
import datetime

# Unix timestamp in milliseconds
timestamp_ms = 1564088876715

# Convert to datetime object
timestamp = datetime.datetime.fromtimestamp(timestamp_ms / 1000.0)

# Format as string
formatted_time = timestamp.strftime('%Y-%m-%d %H:%M:%S')

print(formatted_time)
```

```python
def format_seconds(seconds):
    # Split seconds into minutes and seconds
    minutes, seconds = divmod(seconds, 60)

    # Format as string
    formatted_time = '{:d}:{:06.3f}'.format(int(minutes), seconds)

    return formatted_time

# Example usage
seconds = 513.39233
formatted_time = format_seconds(seconds)
print(formatted_time)
```
