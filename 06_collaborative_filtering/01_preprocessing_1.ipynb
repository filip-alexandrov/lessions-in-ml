```python
import pandas as pd 
df = pd.read_csv("./data/ratings.csv")
```

```python
df
```

```python
# user ids are ordered sequenctially from 1..128493
# with no missing numbers
# movie ids are itegers from 1..131262
# not all movie ids appear 
# there are only 26744 movie ids

# make user ids go from 0..N-1
df['userId'] = df["userId"] - 1
df
```

```python
# create unique mapping for movie ids 
unique_movie_ids = set(df['movieId'].values)
unique_movie_ids
```

```python
movie2idx = {}
count = 0 
for movie_id in unique_movie_ids: 
    movie2idx[movie_id] = count
    count += 1 

movie2idx
```

```python
# add them to the data frame, map each row to the new movie_id
df['movie_idx'] = df.apply(lambda row: movie2idx[row['movieId']], axis=1)
df
```

```python
df = df.drop(columns=['timestamp'])
```

```python
df.to_csv('./data/edited_ratings.csv')
```
